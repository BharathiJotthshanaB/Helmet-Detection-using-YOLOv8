```python
%%writefile merged_data.yaml
path: /content/Merged
train: images/train
val: images/val
test: images/test

names:
  0: Helmet
  1: No_Helmet
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

model.train(
    data="merged_data.yaml",
    epochs=80,
    imgsz=640,
    batch=16,
    name="helmet_final_strong"
)
```
